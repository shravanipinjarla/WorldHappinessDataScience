## Part 2: Eveluation on the HALE metric

```python
# grab the HALE score in all years
all_HALE=pd.merge(pd.merge(pd.merge(pd.merge(C2015['HALE'],C2016['HALE'],how='outer',on='HALE'),C2017['HALE'],how='outer',on='HALE'),C2018['HALE'],how='outer',on='HALE'),C2019['HALE'],how='outer',on='HALE')
all_list=all_HALE['HALE']
all_list
```

```python
#find the lowerq and upperq of all HALE scores
quartiles = all_list.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
lowerq 
iqr = upperq-lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
```

```python
#Plot
fig1, ax0 = plt.subplots()

ax0.set_ylabel('HALE Score')
ax0.boxplot(all_list)
ax0.hlines(y=lowerq, xmin=0.8, xmax=1.2)
ax0.hlines(y=upperq, xmin=0.8, xmax=1.2)
plt.show()
```

```python
# Take year 2019 as example
#Calculate the correlation coefficient for HALE and GDP per Capita

# Identify x and y axises
x_axis1 = C2019['HALE']
y_axis1 = C2019['Economy (GDP per Capita)']

#Plot the line
plt.scatter(x_axis1, y_axis1 ,marker="o")

#set labels and title
plt.xlabel('Healthy Life Expectancy Score')
plt.ylabel('GDP per Capita Score')
plt.title('Correlation of Health and Economy | 2019')

correlation = st.pearsonr(x_axis1,y_axis1)
print(f"The correlation between HALE and GDP per Capita is {round(correlation[0],2)}")

# Calculate the linear regression model
p_slope, p_int, p_r, p_p, p_std_err = st.linregress(x_axis1, y_axis1)
p_fit = p_slope * x_axis1 + p_int

#Plot the line
plt.scatter(x_axis1, y_axis1 ,marker="o")
plt.plot(x_axis1,p_fit,"--",linewidth=3, color='blue')

#plt.show()
plt.savefig('Correlation of Health and Economy.png', dpi=300, bbox_inches='tight')
```
